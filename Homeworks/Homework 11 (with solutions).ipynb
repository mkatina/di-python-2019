```python
class Plane:
    nb_of_planes = 0
    planes_map   = {}

    def __init__(self,
                 model,
                 price,
                 color,
                 speed,
                 max_fuel,
                 fuel_consommation,
                 max_passenger):

        self.model = model
        self.price = price
        self.color = color
        self.speed = speed
        self.max_fuel = max_fuel
        self.fuel_consommation = fuel_consommation
        self.max_passenger = max_passenger

        self.id = Plane.nb_of_planes
        Plane.nb_of_planes += 1

        Plane.planes_map[self.id] = self

        self.current_passengers = 0
        self.current_fuel       = 0
        self.current_airport    = None

    def add_passenger(self, nb_of_passengers):
        if (self.current_passengers + nb_of_passengers) <= self.max_passenger:
            self.current_passengers += nb_of_passengers
            print("{} passengers entered the plane".format(nb_of_passengers))
            return True
        else:
            print("Sorry, too much passengers")
            return False

    def calculate_needed_fuel(self, nb_of_kms):
        return nb_of_kms*self.fuel_consommation

    def refill_fuel(self):
        self.current_fuel = self.max_fuel

    def fly(self, dst_airport, distance):
        needed_fuel = self.calculate_needed_fuel(distance)
        if self.current_fuel < needed_fuel:
            self.refill_fuel()

        self.current_fuel -= needed_fuel
        print("Flying from {} to {}".format(self.current_airport, dst_airport))

        if self.current_airport is not None:
            self.current_airport.plane_out(self)

        dst_airport.plane_in(self)

        self.current_airport = dst_airport

    def __repr__(self):
        return "{} {}".format(self.color, self.model)


    def query(self, id):
        return Plane.planes_map[id]

class Airport:

    def __init__(self, name, capacity):
        # Arguments attributes
        self.name = name
        self.capacity = capacity

        # Default attributes
        self.current_planes_nb = 0

        self.planes = []

    def plane_in(self, plane_obj):
        if (self.current_planes_nb + 1) < self.capacity:
            self.current_planes_nb += 1
            self.planes.append(plane_obj)
            return True
        else:
            return False

    def plane_out(self, plane_obj):
        if (self.current_planes_nb - 1) >= 0:
            self.current_planes_nb -= 1
            self.planes.remove(plane_obj)
            return True
        else:
            return False

    def describe(self):
        print("Airport {}: {}/{} planes".format(self.name,
                                                self.current_planes_nb,
                                                self.capacity))

    def ui(self):

        print("Hey, welcome to {} Airport".format(self.name))

        while True:
            print("\n\n")
            print("What do you want to do? (Choose a number)")
            print("0) Exit")
            print("1) Add planes")
            print("2) Remove planes")
            print("3) Get airport status")

            choice = input(">>> ")
            if choice not in ["0", "1", "2", "3"]:
                print("Please enter 0, 1 or 2")
                continue

            if choice == "0":
                print("Goodbye!")
                break

            elif choice == "1":  # PLANE IN
                nb_of_planes = input("How many planes do you wanna add?\n>>> ")
                try:
                    nb_of_planes = int(nb_of_planes)
                except:
                    print("wrong number of planes")
                    continue

                if self.plane_in(nb_of_planes):
                    print("{} planes entered {}".format(nb_of_planes, self.name))
                else:
                    print("Sorry, no more places available")

            elif choice == "2":  # PLANE OUT
                nb_of_planes = input("How many planes do you wanna remove?\n>>> ")
                try:
                    nb_of_planes = int(nb_of_planes)
                except:
                    print("wrong number of planes")
                    continue

                if self.plane_out(nb_of_planes):
                    print("{} planes left from {}".format(nb_of_planes, self.name))
                else:
                    print("You have no planes in your airport")

            elif choice == "3":
                self.describe()

    def __repr__(self):
        return self.name

plane1 = Plane(model="Boeing 747",
               price=24000,
               color="Blue",
               speed=2400,
               max_fuel=4000,
               fuel_consommation=100,
               max_passenger=250)

plane2 = Plane(model="Boeing 7200",
               price=24000,
               color="yellow",
               speed=2400,
               max_fuel=4000,
               fuel_consommation=100,
               max_passenger=250)

plane3 = Plane(model="random 747",
               price=24000,
               color="red",
               speed=2400,
               max_fuel=4000,
               fuel_consommation=100,
               max_passenger=250)

plane4 = Plane(model="Boeing 737",
               price=24000,
               color="gray",
               speed=2400,
               max_fuel=4000,
               fuel_consommation=100,
               max_passenger=250)

plane5 = Plane(model="airbus 350",
               price=24000,
               color="black",
               speed=2400,
               max_fuel=4000,
               fuel_consommation=100,
               max_passenger=250)

airport1 = Airport("Ben Gurion", 50)
airport2 = Airport("Eilat", 3)
airport3 = Airport("New York", 100)

plane1.fly(airport1, 20)
plane2.fly(airport2, 10)
plane3.fly(airport2, 4)
plane4.fly(airport3, 1)
plane5.fly(airport1, 20)

plane3.fly(airport1, 3)
print(airport1.planes)
plane4.fly(airport1,2)
print(airport1.planes)
```
